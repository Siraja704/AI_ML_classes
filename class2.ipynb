```python
import numpy as np 


data = np.arange(1,101).reshape(10,10)
print(data)
```

```python
np.random.seed(42)
np.random.shuffle(data)
print(data)
```

```python
train_split = int(0.7 * len(data))
val_split = int(0.85 * len(data))


```

```python
train ,val, test = np.split(data, [train_split, val_split])

train 

val

test
```
